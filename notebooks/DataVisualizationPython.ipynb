## Plotting Directly with `pyplot`

We have seen some of the basics of plotting with `pandas`. Many more details and examples are given in the `pandas` documentation (https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html). The `pandas` plotting options are used for working directly with data frames. The `pandas` plotting is based on `mathplotlib` and can do many of the same things. However, `pandas` plotting cannot do everything that `matplotlib` can do and thus there may be times when we want to use `matplotlib` directly. I generally do all of my plotting directly with Pyplot rather than using Pandas plotting. 

All plotting functions  in `matplotlib` take numpy arrays as their input. The documentation for `matplotlib` says that you should convert `pandas` data frames to numpy arrays before plotting, just to be safe. The `pandas` documentation says that data frames act as numpy arrays and thus there is no need to convert. I am going  to take the approach that we will stick with `pandas` data frames unless we encounter a problem. Generally, if there is a problem with a plot, we should be able to tell and thus there shouldn't be any danger in doing this. 


See the `pyplot` tutorial for the basics: 

https://matplotlib.org/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py



Next, we will make some plots from the College Scorecard Data

First import the necessary modules and read in the data. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


collegedata=pd.read_csv("../data/MERGED2015_16_PP.csv",na_values=['NULL','PrivacySuppressed'])

collegedata.shape
```

We will start by making a histogram of the variable HIGHDEG. Recall that this indicates the highest degree awarded at the institution:

0	Non-degree-granting  
1	Certificate degree  
2	Associate degree  
3	Bachelor's degree  
4	Graduate degree  

Note how I specify the bins. The default bins don't align with the values very well. 

```python
plt.hist(collegedata.HIGHDEG,bins=[-0.5,0.5,1.5,2.5,3.5,4.5])
```

Next, I will add labels:

```python
plt.hist(collegedata.HIGHDEG,bins=[-0.5,0.5,1.5,2.5,3.5,4.5])
plt.xlabel("Highest Degree Variable")
plt.title("Distribution of Highest Degree Awarded Across Institutions")
```

A histogram is not the best way to display this data. Among other issues, there is not a way to change from numerical values to descriptive labels. A bar chart is better. For a bar chart, we need to directly provide the counts. Below, I use the function `value_counts` to count up the values in the `HIGHDEG` column. I then make the bar chart. 

```python
plt.figure(figsize=(10,10))
HDcounts=collegedata.HIGHDEG.value_counts()

plt.bar(["Non-degree","Certificate","Associate's Degree","Bachelor's Degree", "Graduate Degree"],HDcounts[[0,1,2,3,4]])
plt.title("Types of Degrees Awarded")
```

I am surprised to see that the largest number (2500) is for institutions that grant certificates. To quote from https://www.bestcolleges.com/features/online-certificate-programs/:

_Certificate programs are available in specialized technical fields; they usually include little to no formal coursework, and instead focus on practicum (or hands-on) training. Professions that may earn certification include plumbers, automotive mechanics, electricians, IT managers, paralegals, and healthcare administrators. Certifications are also widely available in different areas of business management and administration. In most cases, the only prerequisite for a certificate will be a high school diploma or GED. Certificates generally span 20 to 30 credits in length, and most can be earned within one academic year or less._

The second most frequent type of institution are graduate degree granting ones like UGA at about 2000. According to this data, there are fewer than 1000 bachelor's-only institutions. 


Next, I will demonstrate a scatter plot. I will plot the Annual Net Price against the Annual Cost of Attendance seperately for public and private institutions. The net price is the cost of attendance minus financial aid. Thus, this will give us an idea of how financial aid varies according to the cost of the school. I am particularly interested in seeing how this compares between public and private institutions. 

Note that I use the option `figsize=(10,10)`. This specifies the size of the figure. The default size in Jupyter Notebook is too small and the figure is squashed. This fixes that. 

Also note the use of the `subplot` command. The `21` in `plt.subplot(211)` specifies that we should arrange the plots in a 2x1 pattern. The second `1` specifies that the plot should go in the first postion. The second `2` in `plt.subplot(212)` specifies that the figure should go in the second position. 

```python

plt.figure(figsize=(10,10))
plt.subplot(211)
plt.scatter(collegedata.COSTT4_A,collegedata.NPT4_PUB)
plt.title("public institutions")
plt.xlabel("Annual Cost of Attendance")
plt.ylabel("Annual Net Price (Cost of Attenandance - Financial Aid")
plt.subplot(212)
plt.scatter(collegedata.COSTT4_A,collegedata.NPT4_PRIV)
plt.title("private institutions")
plt.xlabel("Annual Cost of Attendance")
plt.ylabel("Annual Net Price (Cost of Attenandance - Financial Aid")
```

We see that the net cost increases linearly with the the total cost for public institutions. Eyeballing it, it appears that the net cost is \\$5,000-\\$10,000 lower than the total cost at all levels of total price. This would tell us that the amount of financial aid at public institution is fairly constant and does not depent on the price of the institution. 


In constrast for private institutions, the net cost appears to flatten out for higher total cost. This would seem to indicate that financial aid is higher at higher cost private institutions. It appears that the net cost tops out at about $30,000 for a large group of institutions, although there are others that go higher. 

In order to make this clearer, I am next going to plot `total cost - net cost`. This will allow us to see directly how the financial depends on total cost: 

```python

plt.figure(figsize=(10,10))
plt.subplot(211)
plt.scatter(collegedata.COSTT4_A,collegedata.COSTT4_A-collegedata.NPT4_PUB)
plt.title("public institutions")
plt.xlabel("Annual Cost of Attendance")
plt.ylabel("Financial aid amount")
plt.subplot(212)
plt.scatter(collegedata.COSTT4_A,collegedata.COSTT4_A-collegedata.NPT4_PRIV)
plt.title("private institutions")
plt.xlabel("Annual Cost of Attendance")
plt.ylabel("financial aid amount")
```

The pattern is clearer now. Financial aid is largely independent of total cost for public institutions and varies from \\$2500-\\$7500 per year. There is a set of institutions in the upper right that break this pattern. 

For private institutions, there is an approximately linear relationship between financial aid amount and the total cost. Institutions that have a total annual cost of \\$30,000 give around \\$10,000 per year in financial aid.Those that have \\$40,000 per year total cost give around \\$20,000 in financial aid. Those that have \\$60,000 per year total cost give around \\$30,000 in financial aid. of course, there is a lot of scatter and also significant deviations from the linear pattern. For example, there is a band of points at the bottom showing a group of institutions where financial aid stays contant at a few thousand dollars per year regardless of price. It would be interesting to figure out what those institutions are, but we won't do that here. It would also be interesting to see how these patterns depend on family income level, but we also will not do that here. 


The last thing that I will go is to make a plot with private and public institutions together in plot with different colors for the variable `ICLEVEL`.  


```python
plt.figure(figsize=(10,10))
colors=pd.Series(["blue","green","red"])  #specify the colors that we want


plt.subplot(211)
plt.scatter(collegedata.COSTT4_A,collegedata.COSTT4_A-collegedata.NPT4_PUB, c=colors[collegedata.ICLEVEL-1])
plt.title("public institutions")
plt.xlabel("Annual Cost of Attendance")
plt.ylabel("Financial aid amount")
plt.subplot(212)
plt.scatter(collegedata.COSTT4_A,collegedata.COSTT4_A-collegedata.NPT4_PRIV,c=colors[collegedata.ICLEVEL-1])
plt.title("private institutions")
plt.xlabel("Annual Cost of Attendance")
plt.ylabel("financial aid amount")
```

Note that there is no legend. This is a hassle to do. We have to make separate plots for each value of ICLEVEL and specify the color for each one along with the label. We can then use the plt.legend to make the legend. 

I show how to do this below. In order to make seperate plots for each value of `ICLEVEL`, we must make subsets of the data by `ICLEVEL`. We have not yet learned how to do this (coming soon!). I demonstrate the procedure below so that you will know how to do it in the future. The syntax `collegedata.COSTT4A_A.loc[collegedata.ICLEVEL==1]` extracts values from `collegedata.COSTT4A_A` where the corresponding value of `ICLEVEL` is equal to 1. 

Note that the `alpha` setting controls the transperancy of the points (0 means invisible, 1 is fully opaque). We can use this if there is a problem with points obscuring other points. I ended up leaving `alpha=1` for all of the plots, but you can try tweaking it. 

```python
plt.figure(figsize=(10,10))  #specify the figure size

plt.subplot(211)  #make two subplots in a 2x1 arrangement
#make a seperate plot for each value of ICLEVEL. Take a subset for each ICLEVEL and plot it seperately. 
plt.scatter(collegedata.COSTT4_A.loc[collegedata.ICLEVEL==1], collegedata.COSTT4_A.loc[collegedata.ICLEVEL==1]-collegedata.NPT4_PUB.loc[collegedata.ICLEVEL==1],color="black",label="4-year",alpha=1)
plt.scatter(collegedata.COSTT4_A.loc[collegedata.ICLEVEL==2], collegedata.COSTT4_A.loc[collegedata.ICLEVEL==2]-collegedata.NPT4_PUB.loc[collegedata.ICLEVEL==2],color="blue", label="2-year",alpha=1)
plt.scatter(collegedata.COSTT4_A.loc[collegedata.ICLEVEL==3], collegedata.COSTT4_A.loc[collegedata.ICLEVEL==3]-collegedata.NPT4_PUB.loc[collegedata.ICLEVEL==3],color="green", label="less than 2-year",alpha=1)
plt.xlabel("annual cost of attendance")
plt.ylabel("financial aid")
plt.title("public institutions")
plt.legend()


plt.subplot(212)
plt.scatter(collegedata.COSTT4_A.loc[collegedata.ICLEVEL==1], collegedata.COSTT4_A.loc[collegedata.ICLEVEL==1]-collegedata.NPT4_PRIV.loc[collegedata.ICLEVEL==1],color="black",label="4-year",alpha=1)
plt.scatter(collegedata.COSTT4_A.loc[collegedata.ICLEVEL==2], collegedata.COSTT4_A.loc[collegedata.ICLEVEL==2]-collegedata.NPT4_PRIV.loc[collegedata.ICLEVEL==2],color="blue", label="2-year",alpha=1)
plt.scatter(collegedata.COSTT4_A.loc[collegedata.ICLEVEL==3], collegedata.COSTT4_A.loc[collegedata.ICLEVEL==3]-collegedata.NPT4_PRIV.loc[collegedata.ICLEVEL==3],color="green", label="less than 2-year",alpha=1)
plt.xlabel("annual cost of attendance")
plt.ylabel("financial aid")
plt.title("private institutions")
plt.legend()




```
